```
import networkx as nax

g = nax.DiGraph()

g.add_node(1)
g.add_node(2)
g.add_node(3)                                                                                                                     
g.add_node(4)
g.add_node(5)

g.add_node(6)
g.add_node(7)
g.add_node(8)


g.add_edge(1,2)
g.add_edge(2,1)
g.add_edge(2,3)
g.add_edge(3,1)
g.add_edge(3,4)
g.add_edge(3,5)
g.add_edge(4,1)
g.add_edge(4,5)
g.add_edge(5,1)
g.add_edge(5,2)
g.add_edge(5,4)

g.add_edge(4,6)
g.add_edge(4,7)
g.add_edge(4,8)
g.add_edge(6,4)
g.add_edge(6,7)
g.add_edge(6,8)
g.add_edge(7,4)
g.add_edge(7,6)
g.add_edge(7,8)
g.add_edge(8,4)
g.add_edge(8,6)
g.add_edge(8,7)


nax.draw_networkx(g)

pr=nax.pagerank(g,alpha=0.85)
print(pr)
```

```
import networkx as nx

g=nx.DiGraph()

nlist=(1,2,3,4,5,6)
g.add_nodes_from(nlist)

elist = [(2,1), (1,3), (3,2), (3,4),(3,5),(4,6),(6,5)]
g.add_edges_from(elist)

nx.draw_networkx(g)

pr=nx.pagerank(g)
pr=sorted(pr.items(),key=lambda x:x[1],reverse=True)
print (pr)
```

```
g1=g.to_undirected()

nx.draw_networkx(g1)

Cliques=[]
for c in nx.find_cliques(g1):
    Cliques.append(set(c))
print(Cliques)
```

# Complex Network/ Social Network

12.17.2020

```
import networkx as nx
import matplotlib.pyplot as plt

lists = []

min=10

max=50

for i in range(min, max):
   print(i)
   cg = nx.cycle_graph(i)
   lists.append(nx.average_shortest_path_length(cg))

   nx.draw_circular(cg, node_size=10, node_color='red')
   print("L =", nx.average_shortest_path_length(cg))

print(lists)
```
